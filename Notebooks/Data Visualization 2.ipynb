```python
# Sequentil color palette
## A sequential color palette is a gradient of colors typically used to represent ordered data that progresses from low to high values. 
##These palettes are particularly useful in visualizing data such as heatmaps, choropleth maps, or any continuous variable.
new2_palette = sns.color_palette("Greens")
sns.palplot(new2_palette)
plt.show()
```

```python
new2_palette = sns.color_palette("light:b")
sns.palplot(new2_palette)
plt.show()
```

```python
# # Divergence color palette
# ## A divergent color palette is used to represent data with a meaningful midpoint, typically where the data diverges in opposite directions
# (e.g., positive vs. negative values or above/below a threshold). 
# Diverging palettes are symmetric, with contrasting colors on either side of the midpoint.
new3_palette = sns.color_palette("BrBG", 9)
sns.palplot(new3_palette)
plt.show()
```

### 1. hue sets up the categorical seperation between an entries of the datasets

A hexbin plot is a type of visualization used to display the relationship between two continuous variables. 
It is particularly useful when dealing with large datasets, 
as it groups points into hexagonal bins and uses color (or intensity) to represent the number of points within each bin. 
This makes it an alternative to scatter plots, which can become cluttered with many overlapping points.

```python
df = sns.load_dataset('iris')
sns.jointplot( x = 'petal_length', y = 'petal_width', data = df,kind='hex')
plt.show()
```

### stripplot()
<p>A stripplot displays individual data points along a categorical axis, 
showing the distribution of values. Each point represents one observation,
with slight random jitter added horizontally to prevent overlap. 
It's useful for seeing the spread and density of data across categories</p>

```python
sns.stripplot(x = "species", y="petal_length",palette="Set1", data=df, jitter = True)
plt.show()
```

<p>A violin plot combines a box plot with a kernel density estimation (KDE) plot, showing:<br>
    
<b>Distribution shape</b> - The width represents how many data points occur at each value<br>
<b>Center line</b> - Shows the median<br>
<b>Box</b> - Displays quartiles (IQR)<br>
<b>Whiskers</b> - Extend to min/max values<br>
<b>Symmetry</b> - Both sides mirror each other, making it resemble a violin<br>

Key benefits:
Shows full distribution shape unlike box plots
Reveals multimodal distributions
Easy comparison between groups
More informative than basic histograms
</p>
