## Bar Plot

```python
x1 = ['apple', 'banana','chiku','dragon','kiwi']
y1 = [23,45,33,12,52]
```

```python
plt.bar(x1,y1)
plt.xlabel('fruits')
plt.ylabel('count')
plt.title('Count of fruits')
plt.show()
```

```python
plt.barh(x1,y1)
plt.xlabel('fruits')
plt.ylabel('count')
plt.title('Count of fruits')
plt.show()
```

## Histogram

```python
mu = 200 # mean of the distribution
sigma = .15 # standard deviation of the distribution
x = mu + sigma + np.random.randn(50000)

num_bins = 40

plt.hist(x,num_bins, facecolor='red', alpha=0.5)
plt.show()
```

```python
import pandas as pd
# creating a sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan","Mike"],
    "Age": [24, 27, 22, 29, 25,23],
    "Gender": ["F", "M", "M", "F", "M","M"],
    "Scores": [88, 92, 95, 85, 90,44]
}
data =pd.DataFrame(data)
data
```

## Boxplots

```python
#box plots works on oly numerical records
# In this case, we can generate boxplots for the column Age and Score

plt.boxplot(data['Age'])
plt.title('boxplot for Age column')
plt.show()
```

```python
plt.boxplot(data['Scores'])
plt.title('boxplot for Scores column')
plt.show()
```

## PIE CHART

```python
# Generating a pie chart
# Creating data for pie chart
d= {'Desserts':["Ice Cream", "Cake", "Cookies", "Brownies", "Pie"],
    'Preferences' : [7.0, 2.5, 8.0, 6.5, 9.0]
}
c_df = pd.DataFrame(d)
c_df
```

```python
# Generating a pie chart
plt.pie(c_df['Preferences'],labels= c_df['Desserts'],startangle=90)
plt.show()
```

```python
# To generate a donut chart, you insert a circle inside a pie chart
# adding circle
plt.pie(c_df['Preferences'],labels= c_df['Desserts'],startangle=90)

circle = plt.Circle((0,0),0.7,color='white')
p = plt.gcf()
p.gca().add_artist(circle)

plt.show()
```
