```python
import pandas as pd
# creating a sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan","Mike"],
    "Age": [24, 27, 22, 29, 25,23],
    "Gender": ["F", "M", "M", "F", "M","M"],
    "Scores": [88, 92, 95, 85, 90,44]
}
data =pd.DataFrame(data)
data
```

## PIE CHART

```python
# Generating a pie chart
# Creating data for pie chart
d= {'Desserts':["Ice Cream", "Cake", "Cookies", "Brownies", "Pie"],
    'Preferences' : [7.0, 2.5, 8.0, 6.5, 9.0]
}
c_df = pd.DataFrame(d)
c_df
```

```python
# Generating a pie chart
plt.pie(c_df['Preferences'],labels= c_df['Desserts'],startangle=90)
plt.show()
```

```python
# To generate a donut chart, you insert a circle inside a pie chart
# adding circle
plt.pie(c_df['Preferences'],labels= c_df['Desserts'],startangle=90)

circle = plt.Circle((0,0),0.7,color='white')
p = plt.gcf()
p.gca().add_artist(circle)

plt.show()
```

## Subplots
### plt.subplots(nrows,ncols)

```python
# we create a sublot of 2 rows and 2 columns and display 4 different plots in each subplot

fig, axs = plt.subplots(2,2)
x = np.arange(1,10)

axs[0,0].plot(x, x**x)
axs[0,0].set_title('square')

axs[0,1].plot(x, np.sqrt(x))
axs[0,1].set_title('square root')

axs[1,0].plot(x, np.exp(x))
axs[1,0].set_title('exp')

axs[1,1].plot(x, np.log10(x))
axs[1,1].set_title('log')

plt.show()
```

# Matplotlib - Grids

```python
fig, axs = plt.subplots(1,2, figsize=(12,4))

x = np.arange(1,19)

axs[0].plot(x, x**3, 'g',lw=2)
axs[0].grid(True)
axs[0].set_title('default demo grid')

axs[1].plot(x, np.sqrt(x), 'r', lw=2)
axs[1].grid(color='b', ls='--', lw=0.25)
axs[1].set_title('custom demo grid')

plt.show()
```
